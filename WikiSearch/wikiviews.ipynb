```python
import requests
import urllib.parse

def find_wikipedia_url(columns):
    all = []
    base_url = 'https://en.wikipedia.org/wiki/'
    for column in columns:
        url = base_url + urllib.parse.quote(column.replace(' ', '_'))
        response = requests.head(url)
        if response.status_code == requests.codes.ok:
            all.append(url)
    return all

webpages = find_wikipedia_url(unique_names)
```
