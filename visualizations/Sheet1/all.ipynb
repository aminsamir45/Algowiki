```python
import pandas as pd
import numpy as np
import sys
```

```python
# Go into above directory and download function
sys.path.append('../../') 
import sheet_to_df

# Extract sheet id - last number of google sheet
sheet_id = 0

data = sheet_to_df.sheet_to_df(sheet_id, 'Sheet1')
```

```python
# cleaning
# Drop rows with missing 'Time Complexity (Average)' or 'Year'
data = data.dropna(subset=['Time Complexity (Average)', 'Year'])

# Remove 'O(' and ')' from 'Time Complexity (Average)'
data['Time Complexity Class'] = data['Time Complexity (Average)'].str.replace('O\(', '', regex=False)
data['Time Complexity Class'] = data['Time Complexity Class'].str.replace(')', '', regex=False)
```

```python
# Check the unique values in 'Time Complexity Class'
unique_complexities = data['Time Complexity Class'].unique()
unique_complexities
```

```python
import plotly.express as px

# Define the list of standard time complexity classes
standard_classes = ["1", "n", "n log n", "n^2", "n^3", "2^n", "n!"]

# Add a new column 'Simplified Complexity Class' that contains the same value as 'Time Complexity Class' for standard classes and 'Other' for non-standard classes
data['Simplified Complexity Class'] = data['Time Complexity Class'].apply(lambda x: x if x in standard_classes else 'Other')

# Create a scatterplot of 'Year' vs 'Simplified Complexity Class'
fig = px.scatter(data, x='Year', y='Simplified Complexity Class',
                 title='Time Complexity Class vs Year of Algorithm Development',
                 labels={'Year':'Year of Algorithm Development', 'Simplified Complexity Class':'Time Complexity Class'})

fig.show()
```
